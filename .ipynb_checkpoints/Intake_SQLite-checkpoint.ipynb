```python
%matplotlib inline
import intake
from intake import open_catalog
```

```python
catalog1 = intake.open_catalog('sqlite1.yml') #reading in new YAML file with updated data
print(list(catalog1))
```

```python
list(catalog1.walkability)
```

Here we see that within our sqlite database no args can be identified. The reason for this data is because our sqlite database has no primary key identified. A primary key in a sql databasde is used to implement data integrity. It is a column or group of columns (composite key in that case) that uniquely identifies rows in a table, allowing you to quiery each table row individually and effectively. Primary keys cannot be defined after a sqlite database has been created and they can't easily be ammended or changed within an existing database. We created our test.db using pandas to convert a CSV file to a DB file, but pandas is unable to simply define a primary key for a sqlite database. To work around this, we found an example SQLite database that has a primary key defined. The data within this example database has no relevance to city information, but it serves as a useful example for demonstrating how we can use the intake catalog source to analyze SQL databases.

```python
catalog2 = intake.open_catalog('sqlite2.yml') #reading in new YAML file with updated data
print(list(catalog2))
```

Within this new catalog, we have an arg called 'chinook'. The arg here is tricky because it is actually in and of itself another catalog that has it's own args. In the next step we demonstrate how this chinook catalog actually holds different data that we can then read and process.

```python
list(catalog2.chinook)
```

There are lots of args such as 'albums', 'artists', etc. hidden within our chinook catalog. We run into an error if we try to read directly the chinook catalog without specifying an arg that we are hoping to parse and analyze. In the next step, we read the data in the chinook catalog regarding artists.

```python
catalog2.chinook.artists.read()
```

If we didn't have a primary key defined in our SQLite database, we would not have been able to locate or read the data in our chinook catalog. Intake-SQL requires that a primary key be defined in a SQLite database. Without a primary key, the intake catalog source isn't able to read through a SQLite database and we get error messages.
