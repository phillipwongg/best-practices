For a local GeoJSON file, the main difference from a remote GeoJSON file is again in the YAML file. Originally, the YAML file used dcat as a driver and was able to gather the datasets from the LA open data source and geohub. The YAML file I used here instead calls on the local GeoJSON files on my computer. To do this, I changed the driver to "geojson" and included the urlpath for the file under args. The urlpath includes a shortcut before the file name that requires my GeoJSON file to be in the same folder and directory as my YAML file. With these changes to my YAML file complete, I can then create a new catalog based on that YAML file. I'm examining the relationship between the walkability of areas in LA to the level of physical activity for those over 18 years old so my two sources of data in the catalog are "walkability_data" and "phys_act_data."

```python
catalog = intake.open_catalog('parking_metro.yml') #reading in new YAML file with updated data
print(list(catalog))
```

After creating the catalog with the two local GeoJSON files, I can now create two dataframes for my walkability data and my physical activity data. Using the catalog increases the reproducability of my code since it requires less boilerplate and is more easily traceable.

```python
walk = catalog.walkability_data
walkability = catalog.walkability_data.read()
```

```python
phys = catalog.phys_act_data
phys_act = catalog.phys_act_data.read()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(16,16))
walkability.plot(ax=ax, color='darkgreen', alpha=0.2, linewidth=1, edgecolor='black')
phys_act.plot(ax=ax, color='pink', alpha=0.5, linewidth=1)
```
